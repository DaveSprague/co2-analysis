```python
df_co2_local = df_co2.copy()
```

```python
df_co2_local = df_co2_local.tz_convert('America/New_York')
df_co2_local.index[0]
```

```python
df_co2_local.index[1]
```

```python
START_TIMESTAMP = '2021-10-25 1:00:00 PM'
END_TIMESTAMP = df_co2_local.index[-1]
```

```python
c0 = df_co2_local.index.to_series().between(START_TIMESTAMP,END_TIMESTAMP)

#create a new dataframe using this index
co2=df_co2_local[c0]
co2[0]
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
  x=co2.index,
  y=co2,
  mode='lines', 
  connectgaps= False,
  name='CO2',
  ))

fig.update_layout(
    title="BAHS Eng Class CO2 Measurements",
)

fig.show()
```

```python
co2.to_csv('co2_export.csv', date_format='%m/%d/%Y %r')
```
